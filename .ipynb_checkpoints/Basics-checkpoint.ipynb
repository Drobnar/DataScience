```python
#Series создание
import pandas as pd

data = pd.Series(["Январь", "Февраль", "Март", "Апрель"],
                 index = ['Первый', "Второй", "Третий", "Четвёртый"])
display(data)
```

```python
#Series доступ к элементам
#loc метка индекса
data.loc[["Первый", "Третий"]]
```

```python
#Series доступ к элементам
#iloc номер индекса
data.iloc[[0, 2]]
```

```python
import pandas as pd
#Группировка данных
#value_counts - подсчет количества
df = pd.read_csv('data_sf.csv')
s = df.Nationality.value_counts()
s
```

```python
#использование условий
df = pd.read_csv('data_sf.csv')
s = df.Nationality.value_counts()
s.loc[s>100]
```

```python
#обращение по индексу
df = pd.read_csv('data_sf.csv')
s = df.Nationality.value_counts()
s.loc['Norway']
```

```python
#value_counts - подсчет количества
#normalize=True - в процентах
df = pd.read_csv('data_sf.csv')
s = df.Nationality.value_counts(normalize=True)
s
```

```python
#unique - уникальные значения
df = pd.read_csv('data_sf.csv')
df.Nationality.unique()
```

```python
#unique - подсчет количества уникальных значений
df = pd.read_csv('data_sf.csv')
df.Nationality.nunique()
```

```python
#Groupby
#сумма по клубам
df = pd.read_csv('data_sf.csv')
df.groupby(['Club']).sum()
```

```python
#сумма по клубам, обращение к определенному клубу
df = pd.read_csv('data_sf.csv')
s = df.groupby(['Club']).sum()
s.loc['Örebro SK']['Wage']
```

```python
#суммирование только по одному столбцу
df = pd.read_csv('data_sf.csv')
s = df.groupby(['Club'])['Wage'].sum()
s.loc['Örebro SK']
```

```python
#сортировка сгруппированного
df = pd.read_csv('data_sf.csv')
s = df.groupby(['Club'])['Wage'].sum().sort_values(ascending=False)
s
```

```python
#выборка сразу по нескольким столбцам
df = pd.read_csv('data_sf.csv')
s = df.groupby(['Club'])['Wage', 'Age', 'ShotPower'].sum().sort_values(['Wage'],ascending=False)
s
```
