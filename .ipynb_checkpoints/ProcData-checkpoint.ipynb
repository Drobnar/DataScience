```python
#columns - изменение названия колонок
import pandas as pd
df = pd.read_csv('log.csv')
df.columns = ['unser_id', 'data', 'bet', 'win']
df
```

```python
#encoding - кодировка
users = pd.read_csv('users.csv', encoding='koi8-r', sep='\t')
users
```

```python
#query - фильтрация
users = pd.read_csv('users.csv', encoding='koi8-r', sep='\t')
users.columns = ['user_id', 'email', 'city']
users.query('city in ["Ижевск", "Воронеж"]')
```

```python
#str.match - строки, которые начинаются
#na = False - исключить ошибку NaN
users = pd.read_csv('users.csv', encoding='koi8-r', sep='\t')
users.columns = ['user_id', 'email', 'city']
users[users.city.str.match('И', na = False)]
```

```python
#str.contains - строки, которые содержат
#na = False - исключить ошибку NaN
users = pd.read_csv('users.csv', encoding='koi8-r', sep='\t')
users.columns = ['user_id', 'email', 'city']
users[users.city.str.contains('еж', na = False)]
```

```python
#apply - преобразование данных (функция)
users = pd.read_csv('users.csv', encoding='koi8-r', sep='\t')
users.columns = ['user_id', 'email', 'city']
users.city = users.city.apply(lambda x: str(x).lower())
users
```
